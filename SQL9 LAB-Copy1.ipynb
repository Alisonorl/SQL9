```python
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import getpass  # to get the password without showing the input
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)
```

```python
data = pd.read_sql_query('SELECT * FROM rentals_may', engine)
data
```

```python
query = '''
        select c.first_name, c.last_name, count(rm.rental_id) as count_may from rentals_may rm
        left join customer c using (customer_id)
        group by c.first_name, c.last_name
        order by count(rm.rental_id) DESC;
        '''
data_may = pd.read_sql_query(query, engine)
data_may.head()
```

```python
data = pd.read_sql_query('SELECT * FROM rentals_june', engine)
data
```

```python
query = '''
        select c.first_name, c.last_name, count(rj.rental_id) as count_june from rentals_june rj
        left join customer c using (customer_id)
        group by c.first_name, c.last_name
        order by count(rj.rental_id) DESC;
        '''
data_june = pd.read_sql_query(query, engine)
data_june.head()
```

```python
data_both=data_may.merge(data_june, right_index=True, left_index=True)
data_both
```

```python
data_both=data_may.merge(data_june, how='inner', on='first_name')
data_both
```

```python
data_both ['difference'] = data_both['count_may'] - data_both['count_june']
data_both
```
